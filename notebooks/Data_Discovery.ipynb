```python
import pandas as pd
import numpy as np

coffee = pd.read_csv("psd_coffee.csv")

population = pd.read_csv(
    "/content/API_SP.POP.TOTL_DS2_EN_csv_v2_3107.csv",
    skiprows=4
)

countries = pd.read_csv(
    "countries-codes (1).csv",
    sep=";"
)
```

```python
print("="*60)
print("Dataset Shapes")
print("="*60)

print(f"Coffee Dataset      : {coffee.shape}")
print(f"Population Dataset  : {population.shape}")
print(f"Country Codes       : {countries.shape}")
```

```python
print("Coffee Columns")
display(coffee.columns)

print("\nPopulation Columns")
display(population.columns)

print("\nCountry Code Columns")
display(countries.columns)
```

```python
print("="*60)
print("Coffee Dataset")
print("="*60)

display(
    coffee[
        ["Country_Name","Country_Code"]
    ].drop_duplicates().head(20)
)
```

```python
print("="*60)
print("Population Dataset")
print("="*60)

display(
    population[
        ["Country Name","Country Code"]
    ].head(20)
)
```

```python
print("="*60)
print("Country Codes Dataset")
print("="*60)

display(
    countries[
        ["LABEL EN","ISO2 CODE","ISO3 CODE"]
    ].head(20)
)
```

```python
coffee_set = set(
    coffee["Country_Name"].dropna().unique()
)

population_set = set(
    population["Country Name"].dropna().unique()
)

countrycode_set = set(
    countries["LABEL EN"].dropna().unique()
)
```

```python
print("="*60)
print("Unique Countries")
print("="*60)

print(f"Coffee Dataset      : {len(coffee_set)}")

print(f"Population Dataset  : {len(population_set)}")

print(f"Country Codes       : {len(countrycode_set)}")
```

```python
# Countries that match exactly by name

exact_matches = coffee_set.intersection(population_set)

print("="*60)
print("Exact Matches")
print("="*60)

print(f"Total Exact Matches: {len(exact_matches)}")
```

```python
# Coffee countries NOT found in Population

coffee_not_population = coffee_set - population_set

print("="*60)
print("Coffee Countries NOT Found in Population")
print("="*60)

for country in sorted(coffee_not_population):
    print(country)
```

```python
# Coffee countries NOT found in Country Codes

coffee_not_countrycodes = coffee_set - countrycode_set

print("="*60)
print("Coffee Countries NOT Found in Country Codes")
print("="*60)

for country in sorted(coffee_not_countrycodes):
    print(country)
```

```python
mismatches = sorted(coffee_not_population)

print(mismatches)
```

```python
def search_population(keyword):
    return population[
        population["Country Name"]
        .str.contains(keyword, case=False, na=False)
    ][["Country Name", "Country Code"]]
```

```python
search_population("viet")
search_population("russ")
search_population("iran")
search_population("vene")
search_population("tur")
search_population("egy")
```
